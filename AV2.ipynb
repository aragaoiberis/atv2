```python
# Define as datas iniciais e finais
start_date = datetime(2021, 1, 5)
end_date = datetime(2021, 1, 30)
num_days = (end_date - start_date).days

# Cria as listas para armazenar as posições dos planetas
dates = []
mercury_ra = []
mercury_dec = []
venus_ra = []
venus_dec = []
mars_ra = []
mars_dec = []
jupiter_ra = []
jupiter_dec = []
saturn_ra = []
saturn_dec = []

# Loop para calcular as posições dos planetas em intervalos de um dia
current_date = start_date
for _ in range(num_days):
    # Calcula as posições dos planetas para a data atual
    current_date_str = current_date.strftime('%Y/%m/%d')
    mercury = ephem.Mercury(current_date_str)
    venus = ephem.Venus(current_date_str)
    mars = ephem.Mars(current_date_str)
    jupiter = ephem.Jupiter(current_date_str)
    saturn = ephem.Saturn(current_date_str)

    # Armazena as posições dos planetas
    dates.append(current_date)
    mercury_ra.append(mercury.ra)
    mercury_dec.append(mercury.dec)
    venus_ra.append(venus.ra)
    venus_dec.append(venus.dec)
    mars_ra.append(mars.ra)
    mars_dec.append(mars.dec)
    jupiter_ra.append(jupiter.ra)
    jupiter_dec.append(jupiter.dec)
    saturn_ra.append(saturn.ra)
    saturn_dec.append(saturn.dec)

    # Incrementa a data em um dia
    current_date += timedelta(days=1)

# Cria o gráfico de Mercúrio
plt.plot(dates, mercury_ra, label='Mercúrio RA')
plt.plot(dates, mercury_dec, label='Mercúrio Dec')

# Cria o gráfico de Vênus
plt.plot(dates, venus_ra, label='Vênus RA')
plt.plot(dates, venus_dec, label='Vênus Dec')

# Cria o gráfico de Marte
plt.plot(dates, mars_ra, label='Marte RA')
plt.plot(dates, mars_dec, label='Marte Dec')

# Cria o gráfico de Júpiter
plt.plot(dates, jupiter_ra, label='Júpiter RA')
plt.plot(dates, jupiter_dec, label='Júpiter Dec')

# Cria o gráfico de Saturno
plt.plot(dates, saturn_ra, label='Saturno RA')
plt.plot(dates, saturn_dec, label='Saturno Dec')

# Configurações do gráfico
plt.xlabel('Data')
plt.ylabel('Posição')
plt.title('Posições dos Planetas em Função do Tempo')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)

# Mostra o gráfico
plt.show()
```

O seguinte código foi gerado no ChatGPT para um padrão de interferência de duas ondas geradas em uma distância d. O código está abaixo

Códigos

b) Faça um fluxograma deste programa.

c) Qual a física relacionada a este código.

```python
import numpy as np
import matplotlib.pyplot as plt

def interference_pattern(distance, wavelength, amplitude):
    # Parâmetros
    separation = distance  # Distância entre as duas ondas
    wave_length = wavelength  # Comprimento de onda
    A = amplitude  # Amplitude

    # Configurações do gráfico
    x = np.linspace(-10 * separation, 10 * separation, 1000)
    y1 = A * np.sin(2 * np.pi * x / wave_length)
    y2 = A * np.sin(2 * np.pi * (x - separation) / wave_length)

    # Soma das duas ondas
    y_sum = y1 + y2

    # Plotar o gráfico
    plt.plot(x, y_sum, label='Interference Pattern')
    plt.plot(x, y1, label='Wave 1')
    plt.plot(x, y2, label='Wave 2')
    plt.xlabel('Distance')
    plt.ylabel('Amplitude')
    plt.title('Interference Pattern between Two Waves')
    plt.legend()
    plt.grid(True)
    plt.show()

# Exemplo de uso
interference_pattern(distance=1, wavelength=1, amplitude=1)
```

QUESTÃO 3  
A sequência de Fibonacci é uma sequência de inteiros **positivos** nos quais cada número é uma soma dos dois anteriores, com os primeiros dois números sendo 1 e 1. Escreva um programa que calcule os 200 primeiros números desta sequência.



```python
def fibonacci(n):
    fib = [1, 1]
    if n == 1 or n==2:
        return fib[:n] #:n -> n primeiros elementos
    #elif n == 2:
        #return fib
    else:
        for i in range(2, n):
            proximo = fib[i-1] + fib[i-2]
            fib.append(proximo)
        return fib
```

```python
n = 200
fibonacci(n)
```

QUESTÃO 4  
Sem usar qualquer função pré estabelecida e usando apenas os comandos de leitura e escrita, além do "for" e "if", após ler um conjunto de números inteiros de um arquivo, classifique-os em ordem crescente.

```python
#para cada i do arquivo, se j for maior que j+1, então trocamos j por j+1
def ord(lista):
    n = len(lista)
    for i in range(n):
        for j in range(0, n-i-1):
            if lista[j] > lista[j+1]:
                troca = lista[j]
                lista[j] = lista[j+1]
                lista[j+1] = troca
    return lista
```

```python
lista = [200, -1, 8, 1.8, 2, 7, np.pi, 4, 4, 354,6027532705, 85, -1, np.e]
ord(lista)
```

Leia do Teclado N e calcule o valor de S:
$$ \frac{1}{N} + \frac{2}{N-1} + \frac{3}{N-2} + ... + \frac{N-1}{2} + \frac{N}{1}$$

*o enunciado me parece confuso, entendi que é para calcular a soma S dado um n fornecido i.e. n = input()*

```python
def Serie(N):
    S = 0
    for k in range (1, N+1):
        soma = k/(N - k + 1)
        S += soma
    return S    
```
